```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Diretório contendo os arquivos CSV
PATH = 'tests/distancia'
```

```python
def scatter_plot(coluna):
    # Inicializa as listas para armazenar dados
    arquivos = []
    valores_y = []

    # Percorre os arquivos CSV no diretório
    for arquivo in os.listdir(PATH):
        if arquivo.endswith(".csv"):
            caminho_arquivo = os.path.join(PATH, arquivo)
            
            # Lê o arquivo CSV
            df = pd.read_csv(caminho_arquivo)
            
            for y in df[coluna]:
                arquivos.append(arquivo)
                valores_y.append(y)  # Exemplo: média dos valores da segunda coluna

    # Cria o gráfico de pontos
    plt.figure(figsize=(10, 6))
    plt.scatter(arquivos, valores_y, color="blue", label="Valores da 2ª coluna")
    plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo X para melhor visualização
    plt.xlabel("Arquivos")
    plt.ylabel("Valores da 2ª Coluna")
    plt.title("Gráfico de Pontos - Valores por Arquivo")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def box_plot(coluna):
    # Inicializa as listas para armazenar dados
    arquivos = []
    valores_y = []
    # Percorre os arquivos CSV no diretório
    for arquivo in os.listdir(PATH):
        if arquivo.endswith(".csv"):
            caminho_arquivo = os.path.join(PATH, arquivo)
            
            # Lê o arquivo CSV
            df = pd.read_csv(caminho_arquivo)
            
            arquivos.append(arquivo)
            valores_y.append(df[coluna])
            
    # Criação do boxplot
    plt.figure(figsize=(8, 6))
    plt.boxplot(valores_y, labels=arquivos, patch_artist=True, boxprops=dict(facecolor="lightblue"))

    # Adicionar detalhes ao gráfico
    plt.xlabel("Distância")
    plt.ylabel("Latência")
    plt.title("Latência/Distancia")
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Exibir o gráfico
    plt.show()
```

```python
scatter_plot('latencia')
```

```python
scatter_plot('fps')
```
